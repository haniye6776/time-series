```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./jena_climate_2009_2016.csv')
df
```

```python
selected_data = df[['T (degC)', 'p (mbar)', 'Tpot (K)', 'Tdew (degC)', 'rh (%)', 'VPmax (mbar)', 'VPact (mbar)', 'VPdef (mbar)',
                        'sh (g/kg)', 'H2OC (mmol/mol)', 'rho (g/m**3)', 'wv (m/s)', 'max. wv (m/s)', 'wd (deg)']]

print("Correlation with (T (degC)):")
print(selected_data.corr()['T (degC)'])
```

```python
columns_to_drop = ['wd (deg)', 'max. wv (m/s)', 'wv (m/s)', 'p (mbar)']
df.drop(columns=columns_to_drop, inplace=True)
df
```

```python
import matplotlib.pyplot as plt


df['Date Time'] = pd.to_datetime(df['Date Time'], format='%d.%m.%Y %H:%M:%S')
df.set_index('Date Time', inplace=True)

time_window_12hr_steps = int(12 * 60 / 10)
time_window_24hr_steps = int(24 * 60 / 10)

features_12hr = []
labels_12hr = []

features_24hr = []
labels_24hr = []

for i in range(len(df) - time_window_24hr_steps):
    feature_window_12hr = df.iloc[i:i + time_window_12hr_steps]['T (degC)'].values
    label_12hr = df.iloc[i + time_window_12hr_steps]['T (degC)']
    features_12hr.append(feature_window_12hr)
    labels_12hr.append(label_12hr)

    feature_window_24hr = df.iloc[i:i + time_window_24hr_steps]['T (degC)'].values
    label_24hr = df.iloc[i + time_window_24hr_steps]['T (degC)']
    features_24hr.append(feature_window_24hr)
    labels_24hr.append(label_24hr)

features_12hr = np.array(features_12hr)
labels_12hr = np.array(labels_12hr)

features_24hr = np.array(features_24hr)
labels_24hr = np.array(labels_24hr)
samples_to_plot = [0, 10000, 20000, 30000, 40000, 50000]

for sample_index in samples_to_plot:
    plt.figure(figsize=(12, 6))
    plt.plot(range(time_window_12hr_steps), features_12hr[sample_index], label='Features (Past Temperature - 12hr)')
    plt.scatter(time_window_12hr_steps, labels_12hr[sample_index], color='red', label='Label (Future Temperature - 12hr)')
    plt.plot(range(time_window_24hr_steps), features_24hr[sample_index], label='Features (Past Temperature - 24hr)')
    plt.scatter(time_window_24hr_steps, labels_24hr[sample_index], color='blue', label='Label (Future Temperature - 24hr)')
    plt.title(f'Temperature Time for Sample {sample_index}')
    plt.xlabel('Time (10-minute steps)')
    plt.ylabel('Temperature (degC)')
    plt.legend()
    plt.show()
    print()
```

***training for 12h prediction***

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler


train_size = int(len(features_12hr) * 0.6)
val_size = int(len(features_12hr) * 0.15)
test_size = len(features_12hr) - train_size - val_size

X12_train = features_12hr[:train_size]
y12_train = labels_12hr[:train_size]
X12_valid = features_12hr[train_size:train_size + val_size]
y12_valid = labels_12hr[train_size:train_size + val_size]
X12_test = features_12hr[train_size + val_size:]
y12_test = labels_12hr[train_size + val_size:]

scaler_12hr = MinMaxScaler()
X12_train_normalized = scaler_12hr.fit_transform(X12_train)
X12_valid_normalized = scaler_12hr.transform(X12_valid)
X12_test_normalized = scaler_12hr.transform(X12_test)
y12_train_normalized = scaler_12hr.fit_transform(y12_train.reshape(-1, 1))
y12_valid_normalized = scaler_12hr.transform(y12_valid.reshape(-1, 1))
y12_test_normalized = scaler_12hr.transform(y12_test.reshape(-1, 1))
```

```python
from tensorflow.keras import layers, models
from sklearn.metrics import mean_squared_error, mean_absolute_error


def create_mlp_model(input_shape):
    model = models.Sequential()
    model.add(layers.Flatten(input_shape=(input_shape,)))
    model.add(layers.Dense(128, activation='relu'))
    model.add(layers.Dense(64, activation='relu'))
    model.add(layers.Dense(1, activation='linear'))

    model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae'])

    return model

input_shape = X12_train_normalized.shape[1]
mlp_model = create_mlp_model(input_shape)
history = mlp_model.fit(X12_train_normalized, y12_train_normalized, epochs=10, batch_size=32, validation_data=(X12_valid_normalized, y12_valid_normalized))

pd.DataFrame(history.history).plot()
```

```python
y12_valid_pred = mlp_model.predict(X12_valid_normalized)
y12_valid_pred_denormalized = scaler_12hr.inverse_transform(y12_valid_pred)
y12_valid_denormalized = scaler_12hr.inverse_transform(y12_valid_normalized)
mse_valid = mean_squared_error(y12_valid_denormalized, y12_valid_pred_denormalized)
mae_valid = mean_absolute_error(y12_valid_denormalized, y12_valid_pred_denormalized)
print(f'Validation - (MSE): {mse_valid}')
print(f'Validation - (MAE): {mae_valid}')

y12_test_pred = mlp_model.predict(X12_test_normalized)
y12_test_pred_denormalized = scaler_12hr.inverse_transform(y12_test_pred)
y12_test_denormalized = scaler_12hr.inverse_transform(y12_test_normalized)
mse_test = mean_squared_error(y12_test_denormalized, y12_test_pred_denormalized)
mae_test = mean_absolute_error(y12_test_denormalized, y12_test_pred_denormalized)
print(f'Test - (MSE): {mse_test}')
print(f'Test - (MAE): {mae_test}')
```

```python
for sample_index in samples_to_plot:
    plt.figure(figsize=(12, 6))
    true_values_range = range(sample_index, sample_index + time_window_12hr_steps)
    plt.plot(true_values_range, y12_test_denormalized[sample_index:sample_index + time_window_12hr_steps], label='True Values', marker='o')
    predicted_values_range = range(sample_index, sample_index + time_window_12hr_steps)
    plt.plot(predicted_values_range, y12_test_pred_denormalized[sample_index:sample_index + time_window_12hr_steps], label='Predicted Values', marker='x')
    plt.title(f'Temperature Prediction - Sample {sample_index} 12-hour')
    plt.xlabel('Time (10-minute steps)')
    plt.ylabel('Temperature (degC)')
    plt.legend()
    plt.show()
    print()
```

***training for 24h prediction***


```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler


train_size = int(len(features_24hr) * 0.6)
val_size = int(len(features_24hr) * 0.15)
test_size = len(features_24hr) - train_size - val_size

X_train = features_24hr[:train_size]
y_train = labels_24hr[:train_size]
X_valid = features_24hr[train_size:train_size + val_size]
y_valid = labels_24hr[train_size:train_size + val_size]
X_test = features_24hr[train_size + val_size:]
y_test = labels_24hr[train_size + val_size:]

scaler_24hr = MinMaxScaler()
X_train_normalized = scaler_24hr.fit_transform(X_train)
X_valid_normalized = scaler_24hr.transform(X_valid)
X_test_normalized = scaler_24hr.transform(X_test)
y_train_normalized = scaler_24hr.fit_transform(y_train.reshape(-1, 1))
y_valid_normalized = scaler_24hr.transform(y_valid.reshape(-1, 1))
y_test_normalized = scaler_24hr.transform(y_test.reshape(-1, 1))
```

```python
from tensorflow.keras import layers, models
from sklearn.metrics import mean_squared_error, mean_absolute_error


def create_mlp_model(input_shape):
    model = models.Sequential()
    model.add(layers.Flatten(input_shape=(input_shape,)))
    model.add(layers.Dense(128, activation='relu'))
    model.add(layers.Dense(64, activation='relu'))
    model.add(layers.Dense(1, activation='linear'))
    model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae'])
    return model

input_shape = X_train_normalized.shape[1]
mlp_model = create_mlp_model(input_shape)
history = mlp_model.fit(X_train_normalized, y_train_normalized, epochs=10, batch_size=32, validation_data=(X_valid_normalized, y_valid_normalized))

pd.DataFrame(history.history).plot()
```

```python
y_valid_pred = mlp_model.predict(X_valid_normalized)
y_valid_pred_denormalized = scaler_12hr.inverse_transform(y_valid_pred)
y_valid_denormalized = scaler_12hr.inverse_transform(y_valid_normalized)
mse_valid = mean_squared_error(y_valid_denormalized, y_valid_pred_denormalized)
mae_valid = mean_absolute_error(y_valid_denormalized, y_valid_pred_denormalized)
print(f'Validation - (MSE): {mse_valid}')
print(f'Validation - (MAE): {mae_valid}')
y_test_pred = mlp_model.predict(X_test_normalized)
y_test_pred_denormalized = scaler_12hr.inverse_transform(y_test_pred)
y_test_denormalized = scaler_12hr.inverse_transform(y_test_normalized)
mse_test = mean_squared_error(y_test_denormalized, y_test_pred_denormalized)
mae_test = mean_absolute_error(y_test_denormalized, y_test_pred_denormalized)
print(f'Test - (MSE): {mse_test}')
print(f'Test - (MAE): {mae_test}')
```

```python
for sample_index in samples_to_plot:
    plt.figure(figsize=(12, 6))
    true_values_range_valid = range(sample_index, sample_index + time_window_12hr_steps)
    selected_data = y_valid_denormalized[sample_index:sample_index + time_window_12hr_steps]
    if len(selected_data) > 0:
        plt.plot(true_values_range_valid, y_valid_denormalized[sample_index:sample_index + time_window_12hr_steps], label='True Values (12hr - Validation)', marker='o')
        predicted_values_range_valid = range(sample_index, sample_index + time_window_12hr_steps)
        plt.plot(predicted_values_range_valid, y_valid_pred_denormalized[sample_index:sample_index + time_window_12hr_steps], label='Predicted Values (12hr - Validation)', marker='x')
        plt.title(f'Validation Temperature Prediction - Sample {sample_index} 12-hour')
        plt.xlabel('Time (10-minute steps)')
        plt.ylabel('Temperature (degC)')
        plt.legend()
        plt.show()
        print()

for sample_index in samples_to_plot:
    plt.figure(figsize=(12, 6))
    true_values_range_test = range(sample_index, sample_index + time_window_12hr_steps)
    plt.plot(true_values_range_test, y_test_denormalized[sample_index:sample_index + time_window_12hr_steps], label='True Values', marker='o')
    predicted_values_range_test = range(sample_index, sample_index + time_window_12hr_steps)
    plt.plot(predicted_values_range_test, y_test_pred_denormalized[sample_index:sample_index + time_window_12hr_steps], label='Predicted Values', marker='x')
    plt.title(f'Test Temperature Prediction - Sample {sample_index} 12-hour')
    plt.xlabel('Time (10-minute steps)')
    plt.ylabel('Temperature (degC)')
    plt.legend()
    plt.show()
    print()
```
